### Popularity Based Recommendation System

Weighted Rating(WR)=[vR/(v+m)]+[mC/(v+m)]
where,
v is the number of votes for the movie;
m is the minimum votes required to be listed in the chart;
R is the average rating of the movie; and
C is the mean vote across the whole report.
Now we find the values of v,m,R,C.

```python
C= movies['vote_average'].mean()
m= movies['vote_count'].quantile(0.9)
```

```python
q_movies = movies.loc[movies['vote_count'] >= m]
```

```python
q_movies
```

```python
q_movies.shape
```

```python
q_movies['score']=np.zeros(len(q_movies))
```

```python
q_movies=q_movies.reset_index()
```

```python
for i in range(len(q_movies)):
    R = q_movies['vote_average'][i]
    v = q_movies['vote_count'][i]
    q_movies['score'][i] = (v / (v + m) * R) + (m / (m + v) * C)
```

```python
q_movies.head(20)
```

```python
q_movies=q_movies.sort_values(by='score',ascending=False)
```

```python
q_movies = q_movies.reset_index(drop=True)
q_movies
```

```python
q_movies=q_movies[q_movies.index<12]
```

```python
q_movies
```

```python
drop=['vote_average','vote_count','score']
q_movies=q_movies.drop(drop,axis=1)
q_movies
```

```python
pickle.dump(q_movies.to_dict(),open('popular.pkl','wb'))
```
